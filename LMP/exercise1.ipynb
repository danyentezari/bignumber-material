```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Construct lines
# x > 0
x = np.linspace(0, 20, 2000)
# y >= 2
y1 = (x*0) + 2
# 2y <= 25 - x
y2 = (25-x)/2.0
# 4y >= 2x - 8 
y3 = (2*x-8)/4.0
# y <= 2x - 5 
y4 = 2 * x -5

# Make plot
plt.plot(x, y1, label=r'$y\geq2')
plt.plot(x, y2, label=r'$2y\leq25-x')
plt.plot(x, y3, label=r'$4y\geq 2x - 8')
plt.plot(x, y4, label=r'$y\leq 2x-5')

plt.xlim((0, 16))
plt.ylim((0, 11))

plt.xlabel(r'$x')
plt.ylabel(r'$y')

# Fill feasible region
y5 = np.minimum(y2, y4)
y6 = np.maximum(y1, y3)
plt.fill_between(x, y5, y6, where=y5>y6, color='grey', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
x = np.linspace(0, 30, 2000)


# Plot the constraints
f1 = (40 - 2 * x)
f2 = (30 - x)

# Set coordinate limits
plt.xlim((0, 50))
plt.ylim((0, 50))

plt.plot(x, f1, label=r'$y\geq2')
plt.fill_between(x, f1, color='grey', alpha=0.5)

plt.plot(x, f2, label=r'$y\geq2')
plt.fill_between(x, f2, color='grey', alpha=0.5)

plt.plot(10, 20, label=r'(10,20)', marker="o", markersize=5, color='black')

# Mask region outside feasible region
plt.fill_between(x, f1, f2, where=f2>f1, color='white', alpha=1)
plt.fill_between(x, f1, f2, where=f2<f1, color='white', alpha=1)

plt.plot(10, 20, label=r'(10,20)', marker="o", markersize=5, color='black')

p = 80*10 + 70*20
ff = (-((8/7)*x) + (1/70)*p)
plt.plot(x, ff, label=r'(10,20)', marker="o", markersize=1, color='red')

print(f'maximum profit is ${(80*10 + 70*20)/100}')
```
