```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
feature_names = iris.feature_names

# Perform PCA on sepal length and sepal width
sepal_data = data[:, :2]  # Extracting sepal length and sepal width
pca = PCA(n_components=2)
sepal_pca = pca.fit_transform(sepal_data)

# Create a DataFrame for the scatter plot
df = pd.DataFrame(data, columns=feature_names)
df['PC1'] = sepal_pca[:, 0]
df['PC2'] = sepal_pca[:, 1]

# Plotting the scatter plots
plt.figure(figsize=(12, 6))

# Scatter plot of sepal length and sepal width
plt.subplot(1, 2, 1)
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=target)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')

# Scatter plot of PC1 and PC2 after PCA on sepal length and sepal width
plt.subplot(1, 2, 2)
plt.scatter(df['PC1'], df['PC2'], c=target)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PC1 vs PC2 after PCA on Sepal Length and Sepal Width')

plt.tight_layout()
plt.show()
```
