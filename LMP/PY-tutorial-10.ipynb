```python
import matplotlib.pyplot as plt
import numpy as np
```

### Vectors in 2D

```python
fig = plt.figure()
ax = plt.axes()

ax.set_xlim([-15,15])
ax.set_ylim([0,10])

# Create a linear space for a vector
def vec(points, xdisplacement=0, ydisplacement=0):
    x = np.linspace(xdisplacement, points[0] + xdisplacement, 100)
    y = np.linspace(ydisplacement, points[1] + ydisplacement, 100)
    return np.array((x, y))

# Draws a vector on a subplot
def draw(v,  obj, xdisplacement=0, ydisplacement=0, c=False):
    obj.plot(v[0] + xdisplacement, v[1] + ydisplacement, c)


# Function to extract points from vector
def gp(v):
    x = v[0][len(v[0])-1]
    y = v[1][len(v[1])-1]
    return np.array((x,y))




v_1 = vec((0,5))
v_2 = vec((0,5))
v_3 = vec(gp(v_1) + gp(v_2))


draw(v_1, obj=ax, c='blue')
draw(v_2, obj=ax, c='red', ydisplacement=5)
draw(v_3, obj=ax, xdisplacement=5)
```

### Vectors in 3D

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_xlim([0,15])
ax.set_ylim([0,15])
ax.set_zlim([0,15])


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Create a linear space for a vector
def vec3d(points, xdisplacement=0, ydisplacement=0, zdisplacement=0):
    x = np.linspace(xdisplacement, points[0] + xdisplacement, 100)
    y = np.linspace(ydisplacement, points[1] + ydisplacement, 100)
    z = np.linspace(zdisplacement, points[2] + zdisplacement, 100)
    return np.array((x, y, z))

# Draws a vector on a subplot
def draw3d(v,  obj, xdisplacement=0, ydisplacement=0,  zdisplacement=0):
    obj.plot(v[0] + xdisplacement, v[1] + ydisplacement, v[2] + zdisplacement)

# Function to extract points from vector
def gp3d(v):
    x = v[0][len(v[0])-1]
    y = v[1][len(v[1])-1]
    z = v[2][len(v[2])-1]
    return np.array((x,y,z))


v1 = vec3d((10,10,10))
draw3d(v1, obj=ax)

v2 = vec3d((10,10,15))
draw3d(v2, obj=ax)
```

### Cosine Similarity and Polar Coordinate System

```python
#        [ t1,  t2,   t3,  t4 ]
#        [ pets, mammals, aneasthesia, building]
word_1 = [0.5,  0.3, 0.1, 0.1]          # (dog)
word_2 = [0.45, 0.3, 0.05, 0.05]        # (cat)
word_3 = [0.01, 0,   0.3,  0.69 ]       # (hospital)
```

```python
import math


fig = plt.figure()
ax = plt.axes()

ax.set_xlim([0,15])
ax.set_ylim([0,15])

# Create a linear space for a vector
def vec(points, xdisplacement=0, ydisplacement=0):
    x = np.linspace(xdisplacement, points[0] + xdisplacement, 100)
    y = np.linspace(ydisplacement, points[1] + ydisplacement, 100)
    return np.array((x, y))

# Draws a vector on a subplot
def draw(v,  obj, xdisplacement=0, ydisplacement=0, c=False):
    obj.plot(v[0] + xdisplacement, v[1] + ydisplacement, c)


# Function to extract points from vector
def gp(v):
    x = v[0][len(v[0])-1]
    y = v[1][len(v[1])-1]
    return np.array((x,y))



def magnitutude(v):
    return math.sqrt(
        sum(
            [math.pow(x_i,2) for x_i in v]
        )
    )
    
    
def cos_sim(v1, v2):
    numerator = np.dot(v1, v2)
    denominator = magnitutude(v1) * magnitutude(v2)
    return numerator/denominator


csim_12 = cos_sim(word_1, word_2)
csim_13 = cos_sim(word_1, word_3)


v_w2 = vec(
    (
        10 * math.sin(csim_12),
        10 * math.cos(csim_12)
    )
)

v_w3 = vec(
    (
        10 * math.sin(csim_13),
        10 * math.cos(csim_13)
    )
)

v_w1 = vec((10,10))
draw(v_w1, obj=ax, c='black')
draw(v_w2, obj=ax, c='red')
draw(v_w3, obj=ax, c='green')


# Convert to Cartesian points
```
