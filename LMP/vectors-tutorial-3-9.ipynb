### 2D

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a linear space for a vector
def vec(points, xdisplacement=0, ydisplacement=0):
    x = np.linspace(xdisplacement, points[0] + xdisplacement, 100)
    y = np.linspace(ydisplacement, points[1] + ydisplacement, 100)
    return np.array((x, y))

# Draws a vector on a subplot
def draw(v,  obj, xdisplacement=0, ydisplacement=0, c=False):
    obj.plot(v[0] + xdisplacement, v[1] + ydisplacement, c)

# Function to extract points from vector
def gp(v):
    x = v[0][len(v[0])-1]
    y = v[1][len(v[1])-1]
    return np.array((x,y))





fig = plt.figure()
ax = plt.axes()

ax.set_xlim([-15,15])
ax.set_ylim([0,10])

v1 = vec((0,4))
v2 = vec((0,4))
v3 = vec(gp(v1) + gp(v2))

draw(v1, obj=ax)
draw(v2, ydisplacement=4, obj=ax)
draw(v3, xdisplacement=5, obj=ax)

```

```python
import math

def magnitutude(v):
    return math.sqrt(sum([math.pow(x,2) for x in v]))

def cos_sim(v1, v2):
    numerator = np.matmul(v1, v2)
    denominator = magnitutude(v1) * magnitutude(v2)
    return numerator/denominator
    
v1 = np.array((0,1,1,1,0))
v2 = np.array((1,1,0,0,1))
cos_sim(v1, v2)

cos_sim(v1, v2)
```

```python
t1 = np.array((0,1,1,1,0))
t2 = np.array((1,1,0,0,1))
cs1 = 1+cos_sim(t1, t2)

t3 = np.array((1,1,1,1,0))
cs2 = cos_sim(t1, t3)

r = 5
newx = r * math.sin(cs1)
newy = r * math.cos(cs1)

v1 = vec((5,5))
v2 = vec((newx, newy))

newx = r * math.sin(cs2)
newy = r * math.cos(cs2)

v3 = vec((newx,newy))

fig = plt.figure()
ax = plt.axes()

ax.set_xlim([0,15])
ax.set_ylim([0,10])


draw(v1, obj=ax, c='black')
draw(v2, obj=ax, c='red')
draw(v3, obj=ax)
```

### 3D

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_xlim([0,15])
ax.set_ylim([0,15])
ax.set_zlim([0,15])


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# ax.set_xticks([0, 15])
# ax.set_yticks([0, 15])
# ax.set_zticks([0, 15])


# Create a linear space for a vector
def vec3d(points, xdisplacement=0, ydisplacement=0, zdisplacement=0):
    x = np.linspace(xdisplacement, points[0] + xdisplacement, 100)
    y = np.linspace(ydisplacement, points[1] + ydisplacement, 100)
    z = np.linspace(zdisplacement, points[2] + zdisplacement, 100)
    return np.array((x, y, z))

# Draws a vector on a subplot
def draw3d(v,  obj, xdisplacement=0, ydisplacement=0,  zdisplacement=0):
    obj.plot(v[0] + xdisplacement, v[1] + ydisplacement, v[2] + zdisplacement)

# Function to extract points from vector
def gp3d(v):
    x = v[0][len(v[0])-1]
    y = v[1][len(v[1])-1]
    z = v[2][len(v[2])-1]
    return np.array((x,y,z))



v1 = vec3d((10,10,10))
draw3d(v1, obj=ax)

v2 = vec3d((10,10,20))
draw3d(v2, obj=ax)
```
