```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats

# Generate normally distributed data
np.random.seed(123)

# Random normally distributed data
x = np.random.normal(loc=5, scale=1, size=50)

# Y is a linear function of X – with some random and normally distributed residuals
y = 2*x + np.random.normal(loc=0, scale=1, size=50)

# Visualize the data
sns.scatterplot(x=x, y=y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of the data')
plt.show()

# Fit the linear regression model
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()


# Check the residuals vs fitted values plot
residuals = model.resid
y_pred = model.predict(X)
plt.scatter(y_pred, residuals)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted values plot')
plt.show()

# Check the histogram of residuals
sns.histplot(residuals, kde=True)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of residuals')
plt.show()

# Check the normal probability plot
sm.qqplot(residuals, line='r')
plt.title('Normal Probability Plot')
plt.show()

# Check the scatterplot of residuals vs independent variable
plt.scatter(x, residuals)
plt.xlabel('Independent variable')
plt.ylabel('Residuals')
plt.title('Residuals vs Independent variable plot')
plt.show()
```
