```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100) + (2*X)

model = sm.OLS(Y, X).fit()

b_1 = model.params[0]

plt.scatter(X, Y)
plt.plot(X, b_1*X)
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.title("Basic Example of Regression with Statsmodels")
plt.show()
```
